```python
import numpy as np
import csv
import json
import pandas as pd
import matplotlib.pyplot as plot
```

```python
AgeGroupDetails = pd.read_csv("AgeGroupDetails.csv")
AgeGroupDetails.to_json (r'AgeGroupDetails.json')
print(AgeGroupDetails)
AgeGroupDetails
```

```python
#removing the % symbol from the percentage column
percentage=[]
percent=AgeGroupDetails['Percentage']
percent
for i in range(len(percent)):
    percentage.append(float(percent[i][:-1]))
percentage
#AgeGroupDetails["Percentage"]=percentage
```

```python
#Analysing Age Group Details
fig = plot.figure()
ax = fig.add_axes([0.2, 0.2, 1.2, 1.2])
ax.bar(AgeGroupDetails['Sno'],percentage,color="green")
ax.set_xlabel('Age Group',fontsize='20')
ax.set_ylabel('Percentage of Cases',fontsize='20')
ax.set_title('Covid19 Age Group Details (In Percentage )',fontsize='20')
ax.set_xticks(AgeGroupDetails['Sno'])
ax.set_xticklabels(AgeGroupDetails['AgeGroup'])
plot.show()


fig2 = plot.figure()
ax = fig2.add_axes([0.2, 0.2, 1.2, 1.2])
ax.bar(AgeGroupDetails['Sno'],AgeGroupDetails['TotalCases'],color="blue")
ax.set_xlabel('Age Group',fontsize='20')
ax.set_ylabel('Total Number of Cases',fontsize='20')
ax.set_title('Covid19 Age Group Details',fontsize='20')
ax.set_xticks(AgeGroupDetails['Sno'])
ax.set_xticklabels(AgeGroupDetails['AgeGroup'])
plot.show()
```

```python
covid_19_india = pd.read_csv("covid_19_india.csv")
covid_19_india.to_json (r'covid_19_india.json')
covid_19_india
```

```python
#Total Number of confirmed cases State wise
confirmed_cases=covid_19_india.groupby('State/UnionTerritory')['Confirmed'].max()
print(confirmed_cases)
```

```python
#Graphical Representation of Total Confirmed Cases State Wise
confirmed_cases.plot(kind='bar',figsize=(15, 15),color="purple")
plot.xlabel("State/UnionTerritory",fontsize="20")
plot.ylabel("Total Confirmed Cases",fontsize="20")
plot.title("Bar Graph for Total Nuber of Confirmed Cases State Wise",fontsize="20")
```

```python
#Analysis for Delhi
print("Total Number of Confirmed Cases In Delhi = ",confirmed_cases['Delhi'])
delhi_data=covid_19_india.loc[covid_19_india['State/UnionTerritory'] == "Delhi"]
delhi_data.head()
delhi_data.plot(figsize=(15, 10))
```
